# Import library

```python
import pandas as pd
import bqplot
import numpy as np
import ipywidgets
import matplotlib.pyplot as plt
```

# Import Data

```python
data_url = 'https://data.illinois.gov/dataset/9c0ea828-9734-437f-b612-e503d7480c9f/resource/e20d244b-96a2-4848-ba9d-83cb35c57b05/download/cspermits_20191226.csv'

permits = pd.read_csv(data_url, encoding= 'unicode_escape')
```

```python
permits
```

# Data processing

###  Extract Component data

```python
component = permits[['Permit Type','Issue Year','Improvement Type','Estimated Cost','Fee Amount','Permit Status']].copy()
```

```python
component = component.dropna()
```

```python
component
```

# Get the data for left table

```python
left_component= component[['Permit Type','Permit Status','Estimated Cost']].copy()
```

```python
left_component = left_component.pivot_table(values='Estimated Cost', index=['Permit Type'], columns=['Permit Status'], 
                      aggfunc = np.mean, fill_value=0)
```

```python
left_component = left_component.dropna()
```

```python
left_component
```

# Plot left component

```python
myLabel = ipywidgets.Label()
def on_selected(change):
    if len(change['owner'].selected) == 1: # select one bin at a time
        i,j = change['owner'].selected[0]
        v = left_component.iat[i, j]
        myLabel.value = 'Estimated Cost ='+str(v)
        Permit_Type = left_component.index[i]
        Permit_Status = left_component.columns[j]
        
# scales
col_sc = bqplot.ColorScale(scheme='OrRd')
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# axis
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')
x_ax = bqplot.Axis(scale = x_sc,tick_rotate = 15,tick_style={'font-size': 10},offset={'scale':x_sc, 'value':50})
y_ax = bqplot.Axis(scale=y_sc, tick_rotate = 20,orientation='vertical',tick_style={'font-size': 7})

x_label = left_component.index.values.astype('str')
y_label = left_component.columns.values.astype('str')

# marks
heat_map = bqplot.GridHeatMap(color=left_component.values, row=x_label, column=y_label,
                              scales={'color':col_sc,'row':y_sc, 'column':x_sc},
                              interactions = {'click': 'select'}, 
                              anchor_style = {'fill':'grey'}, 
                              selected_style = {'opacity': 1.0}, 
                              unselected_style = {'opacity': 0.8},padding=0.5,stroke='white') 


heat_map.observe(on_selected,'selected')

fig = bqplot.Figure(marks=[heat_map], axes=[col_ax, x_ax, y_ax],fig_margin={'top':100, 'bottom':100, 'left':100, 'right':100})
#fig
myDashboard = ipywidgets.VBox([myLabel,fig])
myDashboard
```
